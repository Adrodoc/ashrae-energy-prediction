```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
s = pd.Series([datetime.now(), datetime(2019, 1, 2, 2)])
np.sin((2 * np.pi) * ((s.dt.dayofyear - 1) * 24 + s.dt.hour) / (24 * 365))
```

```python
timestamp = pd.Series(np.arange(12 * 28)).map(lambda x: datetime(2019, x%12 + 1, x//12 + 1))
hourofyear_sin = np.sin((2 * np.pi) * ((timestamp.dt.dayofyear - 1) * 24 + timestamp.dt.hour) / (24 * 365))
df = pd.DataFrame({"timestamp": timestamp, "hourofyear_sin": hourofyear_sin})
df.sort_values(by="timestamp", inplace=True)
print(df.head())
plt.plot(df.groupby("timestamp").min())
```
